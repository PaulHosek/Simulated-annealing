```python
from charges import Charges
from plotting import *

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
import time
from itertools import combinations

%matplotlib widget
```

```python
# Parameters
methods = ["linear", "exponential_even_spacing", "exponential_0.003"]

n = 12
method = methods[2]
iterations = 100
markov = [5, 10, 50, 100]

# ch = Charges(n)
# start = np.copy(ch.particles)
# print(f"Total energy of the system: {ch.evaluate_configuration()}")
convergence = 65.0
```

```python
res_list = []

#loop over iteration/markov combinations
for m in markov:
    temp = []
    # do combination 10 times
    for j in range(100):
        ch = Charges(n)
        print(j, iterations, m)
        res = ch.iterate_SA_markovergence(0.1, 50, iterations, methods[2], m, convergence, force=0)
        print('')
        temp.append(res)
    res_list.append(temp)
```

```python
fig, ax = plt.subplots()
# plt.plot(markov, [np.mean(r) for r in res_list])
plt.errorbar(markov, [np.mean(r) for r in res_list], [np.std(r) for r in res_list])
ax.set_xscale('log')
ax.set_xlabel('Markov chain length')
ax.set_ylabel('Number of evaluations before convergence to E=65.0')
ax.set_ylim(0,25000)
plt.show()
```
