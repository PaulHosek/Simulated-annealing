```python
# for all forces once + for noforce & all cooling schemes
nr_sim = 100
schedule = schedules[0]
linear_data = np.empty((n_temps*chain_length*n_charges-1,nr_sim))
for i in range(nr_sim):
    my_charge = charges.Charges(n_charges,1,step_size=0.001)
    res = my_charge.iterate_SA_optimize(low_temp, high_temp, n_temps, schedule, chain_length,force,wavy)

    fname = f"{n_charges}_{schedule}_{n_temps}_{chain_length}_{force_name}_{wavy_name}"
    # read energies into np array column
    res_df = pd.read_csv(f"logged_data/{fname}.csv", skiprows=[1]).rename(columns={"Unnamed: 0": "Iterations"})
    linear_data[:,i] = res_df['Potential_energy']

linear_data
np.savetxt("logged_data/linear_data_100runs_wavy_noforce.csv", linear_data, delimiter=",")
```
